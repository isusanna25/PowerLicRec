```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import re
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
import numpy as np
import pandas as pd
import pickle

url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+|www\.\w+\.\w+')

def extract_keywords(text):
    # 去除文本中可能的网址信息
    text = url_pattern.sub('',text)
    # 去除非英文字符
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    # 将文本转换为小写并进行分词
    tokens = word_tokenize(text.lower())
    # 标记每个单词的词性
    tagged_tokens = pos_tag(tokens)
    # 仅选择名词和形容词作为关键词
    keywords = [word for word, pos in tagged_tokens if pos.startswith('N')]
    # 去除标点符号
    keywords = [word for word in keywords if word.isalnum()]
    return keywords

with open('./kmeans.pkl', 'rb') as f:
    kmeans = pickle.load(f)
#     prd_y = model.predict(x)

with open('./rf.pkl', 'rb') as f:
    rf = pickle.load(f)
#     prd_y = model.predict(x)
    
with open('./language.pkl', 'rb') as f:
    languages = pickle.load(f)
#     prd_y = model.predict(x)
    
with open('./year.pkl', 'rb') as f:
    years = pickle.load(f)
#     prd_y = model.predict(x)

licenses = {'MIT License': 0, #写一个没保存的对比一下
 'GNU General Public License v3.0': 1,
 'Apache License 2.0': 2,
 'BSD 3-Clause "New" or "Revised" License': 3,
 'GNU General Public License v2.0': 4,
 'GNU Affero General Public License v3.0': 5,
 'GNU Lesser General Public License v3.0': 6,
 'BSD 2-Clause "Simplified" License': 7,
 'Mozilla Public License 2.0': 8}
licenses = { j:i for i, j in licenses.items()}

# df['word1'] = df.apply(lambda x: extract_keywords(str(x['description'])) if x['description'] != '' else '', axis=1) #输出description中所有的名词
# df['word2'] = df.apply(lambda x: extract_keywords(str(x['readme_text'])) if x['readme_text'] != '' else '', axis=1) #输出readme中所有的名词
# df['word'] = df.apply(lambda x: x['word1'] + x['word2'] if x['word1']!='' and x['word2']!='' else x['word1'] if x['word1']!='' else x['word2'] if x['word2']!='' else '', axis=1)
```
